simple plot 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2, 6, 50)
y1 = x + 3 
y2 = 3 - x        

plt.plot(x,y1,color='red',linestyle='--',label='y1')
plt.plot(x,y2,color='blue',linestyle='--',label='y2')
plt.legend(loc="upper left")
plt.xlabel('random x')
plt.ylabel('mapping')
plt.title('test random plot', color='blue', fontsize=16)
plt.show()
```

correlation plot

```python
import pandas as pd
import numpy as np

rs = np.random.RandomState(0)
df = pd.DataFrame(rs.rand(10, 10))
corr = df.corr()
corr.style.background_gradient(cmap='BrBG')
```
